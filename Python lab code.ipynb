```python
 'Array Reversal program'

arr = [1, 4, 3, 2, 6, 5]

i, j = 0, len(arr) - 1
while i < j:
    arr[i], arr[j] = arr[j], arr[i]
    i += 1
    j -= 1

print("Reversed Array:", arr)
```

```python
'Minimum and maximum elements program in an array'

arr = [1, 4, 3, 5, 8, 6]

minimum = arr[0]
maximum = arr[0]

for x in arr:
    if x < minimum:
        minimum = x
    if x > maximum:
        maximum = x

print("Minimum:", minimum)
print("Maximum:", maximum)
```

```python
'Cyclically Rotate an Array by One Position (clockwise)'

arr = [1, 2, 3, 4, 5]

last = arr[-1]

for i in range(len(arr) - 1, 0, -1):
    arr[i] = arr[i - 1]

arr[0] = last

print("Array after rotation:", arr)
```

```python
def rotate_array(arr):
    if len(arr) > 0:
        last = arr[-1]
        for i in range(len(arr)-1, 0, -1):
            arr[i] = arr[i-1]
        arr[0] = last
    return arr

# User input
arr = list(map(int, input("Enter the array elements separated by space: ").split()))
print("Original array:", arr)
print("Rotated array:", rotate_array(arr))
```

```python
'''Maximum Sum of a Subarray (Kadane's Algorithm) '''

def max_subarray_sum(arr):
    max_current = max_global = arr[0]
    for num in arr[1:]:
        max_current = max(num, max_current + num)
        max_global = max(max_global, max_current)
    return max_global

# User input
arr = list(map(int, input("Enter the array elements separated by space: ").split()))
print("Max subarray sum:", max_subarray_sum(arr))
```

```python
 '''Search Insert Position (binary search, O(log n))'''

def search_insert(nums, target):
    left, right = 0, len(nums) - 1
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return left

# User input
nums = list(map(int, input("Enter the sorted array elements separated by space: ").split()))
target = int(input("Enter the target value: "))
print("Insert position:", search_insert(nums, target))
```

```python
'''Two Sum (return indices of two numbers adding to target)'''

def two_sum(nums, target):
    num_map = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
    return []

# User input
nums = list(map(int, input("Enter the array elements separated by space: ").split()))
target = int(input("Enter the target value: "))
print("Two sum indices:", two_sum(nums, target))
```

```python
'''Minimum Number of Jumps to Reach End of Array'''

def min_jumps(arr):
    n = len(arr)
    if n <= 1:
        return 0
    if arr[0] == 0:
        return -1  # can't move forward
    jumps = 1
    max_reach = arr[0]
    steps = arr[0]
    for i in range(1, n):
        if i == n - 1:
            return jumps
        max_reach = max(max_reach, i + arr[i])
        steps -= 1
        if steps == 0:
            jumps += 1
            if i >= max_reach:
                return -1
            steps = max_reach - i
    return -1

# User input
arr = list(map(int, input("Enter the array elements separated by space: ").split()))
print("Minimum jumps:", min_jumps(arr))
```
