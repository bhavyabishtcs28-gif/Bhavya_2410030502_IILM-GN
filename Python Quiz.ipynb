```python
"Printing sum of odd numbers between 1 to 100"

total = 0

for num in range(1, 101):
    if num % 2 != 0:   # check if number is odd
        total += num

print("Sum of odd numbers from 1 to 100 is:", total)
```

```python
adult = int(input("Are you an adult? (1 if yes, 0 if minor): "))
married = int(input("Are you married? (1 if yes, 0 if single): "))

if adult == 1 and married == 1:
    print("You are an adult and married.")
elif adult == 1 and married == 0:
    print("You are an adult and single.")
elif adult == 0 and married == 1:
    print("You are a minor and married. (This is unusual in most places.)")
elif adult == 0 and married == 0:
    print("You are a minor and single.")
else:
    print("Invalid input! Please enter only 1 or 0.")
```

```python
for num in range(2, 13):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, "is a Prime number")
    else:
        print(num, "is a Composite number")
```

```python
for num in range(100, 1000):
    temp = num
    sum_of_cubes = 0

    while temp > 0:
        digit = temp % 10
        sum_of_cubes += digit ** 3
        temp //= 10

    if sum_of_cubes == num:
        print(num, " is an Armstrong number")
```

```python
lst = [5, 6, 3, 9, 2, 12, 3, 8, 7]

n = len(lst)

for i in range(n - 1):
    if lst[i] > lst[i + 1]:
        lst[i], lst[i + 1] = lst[i + 1], lst[i]   # tuple swap (no temp)

print(lst)
```
