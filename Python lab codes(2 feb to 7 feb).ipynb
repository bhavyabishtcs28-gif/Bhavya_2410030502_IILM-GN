```python
''' Minimum Difference in Chocolate Distribution '''

def min_diff_chocolates(arr, m):
    """
    Calculates the minimum difference in chocolate distribution.
    
    Parameters:
    arr (list): The input array of chocolate packets.
    m (int): The number of students.
    
    Returns:
    int: The minimum possible difference.
    """
    # Sort the array
    arr.sort()
    
    # Initialize the minimum difference
    min_diff = float('inf')
    
    # Iterate over the array with a sliding window of size m
    for i in range(len(arr) - m + 1):
        # Calculate the difference for the current window
        diff = arr[i + m - 1] - arr[i]
        
        # Update the minimum difference
        min_diff = min(min_diff, diff)
    
    return min_diff

def main():
    # Get the array size from the user
    n = int(input("Enter the size of the array: "))
    
    # Get the array elements from the user
    arr = []
    print("Enter elements of the array:")
    for i in range(n):
        num = int(input(f"Enter element {i + 1}: "))
        arr.append(num)
    
    # Get the number of students from the user
    m = int(input("Enter the number of students: "))
    
    # Check if the number of students is valid
    if m > n:
        print("Error: Number of students cannot be more than the number of packets.")
    else:
        # Calculate the minimum difference
        min_diff = min_diff_chocolates(arr, m)
        
        # Print the result
        print(f"The minimum possible difference is: {min_diff}")

if __name__ == "__main__":
    main()

```

```python
''' Smallest Subarray with Sum Greater than X '''

def smallest_subarray(arr, x):
    """
    Finds the smallest subarray with sum greater than the given value.
    
    Parameters:
    arr (list): The input array of integers.
    x (int): The target sum.
    
    Returns:
    int: The length of the smallest subarray, or 0 if no such subarray exists.
    """
    # Initialize the window boundaries
    left = 0
    current_sum = 0
    min_length = float('inf')
    
    # Iterate over the array
    for right in range(len(arr)):
        # Add the current element to the window sum
        current_sum += arr[right]
        
        # Shrink the window if the sum is greater than x
        while current_sum > x:
            min_length = min(min_length, right - left + 1)
            current_sum -= arr[left]
            left += 1
    
    # Return the minimum length, or 0 if no such subarray exists
    return min_length if min_length != float('inf') else 0

def main():
    # Get the array size from the user
    n = int(input("Enter the size of the array: "))
    
    # Get the array elements from the user
    arr = []
    print("Enter elements of the array:")
    for i in range(n):
        num = int(input(f"Enter element {i + 1}: "))
        arr.append(num)
    
    # Get the target sum from the user
    x = int(input("Enter the target sum: "))
    
    # Find the smallest subarray
    result = smallest_subarray(arr, x)
    
    # Print the result
    if result == 0:
        print("No such subarray exists.")
    else:
        print(f"The length of the smallest subarray is: {result}")

if __name__ == "__main__":
    main()

```

```python
''' Partitioning Array around a Range '''

def partition_array(arr, a, b):
    """
    Partitions the array around the given range.
    
    Parameters:
    arr (list): The input array.
    a (int): The lower bound of the range.
    b (int): The upper bound of the range.
    
    Returns:
    list: The modified array.
    """
    # Initialize three lists for the three partitions
    smaller = []
    between = []
    greater = []
    
    # Partition the array
    for num in arr:
        if num < a:
            smaller.append(num)
        elif a <= num <= b:
            between.append(num)
        else:
            greater.append(num)
    
    # Combine the partitions
    return smaller + between + greater

def main():
    # Get the array size from the user
    n = int(input("Enter the size of the array: "))
    
    # Get the array elements from the user
    arr = []
    print("Enter elements of the array:")
    for i in range(n):
        num = int(input(f"Enter element {i + 1}: "))
        arr.append(num)
    
    # Get the range from the user
    a = int(input("Enter the lower bound of the range: "))
    b = int(input("Enter the upper bound of the range: "))
    
    # Partition the array
    result = partition_array(arr, a, b)
    
    # Print the result
    print("The modified array is:", result)

if __name__ == "__main__":
    main()
```

```python
''' Minimum Swaps to Bring Elements Together '''

def min_swaps(arr, k):
    """
    Finds the minimum number of swaps required to bring all elements less than or equal to k together.
    
    Parameters:
    arr (list): The input array.
    k (int): The target number.
    
    Returns:
    int: The minimum number of swaps required.
    """
    # Count the number of elements less than or equal to k
    count = sum(1 for num in arr if num <= k)
    
    # Initialize the window boundaries
    left = 0
    bad = 0
    
    # Initialize the minimum swaps
    min_swaps = float('inf')
    
    # Iterate over the array
    for right in range(len(arr)):
        # Add the current element to the window
        if arr[right] > k:
            bad += 1
        
        # Shrink the window if it's larger than the count
        if right - left + 1 > count:
            if arr[left] > k:
                bad -= 1
            left += 1
        
        # Update the minimum swaps
        if right - left + 1 == count:
            min_swaps = min(min_swaps, bad)
    
    return min_swaps

def main():
    # Get the array size from the user
    n = int(input("Enter the size of the array: "))
    
    # Get the array elements from the user
    arr = []
    print("Enter elements of the array:")
    for i in range(n):
        num = int(input(f"Enter element {i + 1}: "))
        arr.append(num)
    
    # Get the target number from the user
    k = int(input("Enter the target number: "))
    
    # Find the minimum swaps
    result = min_swaps(arr, k)
    
    # Print the result
    print(f"The minimum number of swaps required is: {result}")

if __name__ == "__main__":
    main()

```

```python
''' Checking if All Array Elements are Palindromes '''

def is_palindrome(num):
    """
    Checks if a number is a palindrome.
    
    Parameters:
    num (int): The input number.
    
    Returns:
    bool: True if the number is a palindrome, False otherwise.
    """
    return str(num) == str(num)[::-1]

def all_palindromes(arr):
    """
    Checks if all elements in the array are palindromes.
    
    Parameters:
    arr (list): The input array.
    
    Returns:
    bool: True if all elements are palindromes, False otherwise.
    """
    return all(is_palindrome(num) for num in arr)

def main():
    # Get the array size from the user
    n = int(input("Enter the size of the array: "))
    
    # Get the array elements from the user
    arr = []
    print("Enter elements of the array:")
    for i in range(n):
        num = int(input(f"Enter element {i + 1}: "))
        arr.append(num)
    
    # Check if all elements are palindromes
    result = all_palindromes(arr)
    
    # Print the result
    if result:
        print("All elements in the array are palindromes.")
    else:
        print("Not all elements in the array are palindromes.")

if __name__ == "__main__":
    main()
```

```python
''' Calculating the Median of an Array '''

def calculate_median(arr):
    """
    Calculates the median of the input array.
    
    Parameters:
    arr (list): The input array.
    
    Returns:
    float: The median of the array.
    """
    # Sort the array
    arr.sort()
    
    # Calculate the median
    n = len(arr)
    if n % 2 == 1:
        # Odd number of elements
        median = arr[n // 2]
    else:
        # Even number of elements
        median = (arr[n // 2 - 1] + arr[n // 2]) / 2
    
    return median

def main():
    # Get the array size from the user
    n = int(input("Enter the size of the array: "))
    
    # Get the array elements from the user
    arr = []
    print("Enter elements of the array:")
    for i in range(n):
        num = int(input(f"Enter element {i + 1}: "))
        arr.append(num)
    
    # Calculate the median
    median = calculate_median(arr)
    
    # Print the result
    print(f"The median of the array is: {median}")

if __name__ == "__main__":
    main()
```

```python
''' Traversing a Matrix in Spiral Form '''

def spiral_traversal(mat):
    """
    Traverses the input matrix in spiral form.
    
    Parameters:
    mat (list): The input matrix.
    
    Returns:
    list: The elements of the matrix in spiral order.
    """
    result = []
    top = 0
    bottom = len(mat) - 1
    left = 0
    right = len(mat[0]) - 1
    
    while top <= bottom and left <= right:
        # Traverse from left to right
        for i in range(left, right + 1):
            result.append(mat[top][i])
        top += 1
        
        # Traverse from top to bottom
        for i in range(top, bottom + 1):
            result.append(mat[i][right])
        right -= 1
        
        # Traverse from right to left
        if top <= bottom:
            for i in range(right, left - 1, -1):
                result.append(mat[bottom][i])
            bottom -= 1
        
        # Traverse from bottom to top
        if left <= right:
            for i in range(bottom, top - 1, -1):
                result.append(mat[i][left])
            left += 1
    
    return result

def main():
    # Get the matrix dimensions from the user
    n = int(input("Enter the number of rows: "))
    m = int(input("Enter the number of columns: "))
    
    # Get the matrix elements from the user
    mat = []
    print("Enter elements of the matrix:")
    for i in range(n):
        row = []
        for j in range(m):
            num = int(input(f"Enter element ({i + 1}, {j + 1}): "))
            row.append(num)
        mat.append(row)
    
    # Traverse the matrix in spiral form
    result = spiral_traversal(mat)
    
    # Print the result
    print("The elements of the matrix in spiral order are:", result)

if __name__ == "__main__":
    main()
```

```python
''' Searching in a Sorted Matrix '''

def search_matrix(mat, target):
    """
    Searches for the target in the input matrix.
    
    Parameters:
    mat (list): The input matrix.
    target (int): The target to search for.
    
    Returns:
    bool: True if the target is found, False otherwise.
    """
    m, n = len(mat), len(mat[0])
    left, right = 0, m * n - 1
    
    while left <= right:
        mid = (left + right) // 2
        mid_val = mat[mid // n][mid % n]
        
        if mid_val == target:
            return True
        elif mid_val < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return False

def main():
    # Get the matrix dimensions from the user
    m = int(input("Enter the number of rows: "))
    n = int(input("Enter the number of columns: "))
    
    # Get the matrix elements from the user
    mat = []
    print("Enter elements of the matrix (in sorted order):")
    for i in range(m):
        row = []
        for j in range(n):
            num = int(input(f"Enter element ({i + 1}, {j + 1}): "))
            row.append(num)
        mat.append(row)
    
    # Get the target from the user
    target = int(input("Enter the target to search for: "))
    
    # Search for the target
    result = search_matrix(mat, target)
    
    # Print the result
    if result:
        print(f"The target {target} is found in the matrix.")
    else:
        print(f"The target {target} is not found in the matrix.")

if __name__ == "__main__":
    main()
```

```python
''' Finding the Median of a Sorted Matrix '''

def count_less_than_or_equal_to(mat, num):
    """
    Counts the number of elements less than or equal to num in the matrix.
    
    Parameters:
    mat (list): The input matrix.
    num (int): The target number.
    
    Returns:
    int: The count of elements less than or equal to num.
    """
    count = 0
    n, m = len(mat), len(mat[0])
    j = m - 1
    
    for i in range(n):
        while j >= 0 and mat[i][j] > num:
            j -= 1
        count += (j + 1)
    
    return count

def find_median(mat):
    """
    Finds the median of the input matrix.
    
    Parameters:
    mat (list): The input matrix.
    
    Returns:
    int: The median of the matrix.
    """
    n, m = len(mat), len(mat[0])
    low, high = mat[0][0], mat[n - 1][m - 1]
    
    while low < high:
        mid = (low + high) // 2
        count = count_less_than_or_equal_to(mat, mid)
        
        if count < (n * m + 1) // 2:
            low = mid + 1
        else:
            high = mid
    
    return low

def main():
    # Get the matrix dimensions from the user
    n = int(input("Enter the number of rows: "))
    m = int(input("Enter the number of columns: "))
    
    # Check if the dimensions are odd
    if n % 2 == 0 or m % 2 == 0:
        print("Error: The number of rows and columns must be odd.")
        return
    
    # Get the matrix elements from the user
    mat = []
    print("Enter elements of the matrix (in sorted order):")
    for i in range(n):
        row = []
        for j in range(m):
            num = int(input(f"Enter element ({i + 1}, {j + 1}): "))
            row.append(num)
        mat.append(row)
    
    # Find the median
    median = find_median(mat)
    
    # Print the result
    print(f"The median of the matrix is: {median}")

if __name__ == "__main__":
    main()
```

```python
''' Finding the Index of the Row with Maximum 1s '''

def find_max_ones_row(mat):
    """
    Finds the index of the first row that contains the maximum number of 1s.
    
    Parameters:
    mat (list): The input matrix.
    
    Returns:
    int: The index of the first row with maximum 1s, or -1 if no such row exists.
    """
    n, m = len(mat), len(mat[0])
    max_ones = 0
    max_index = -1
    j = m - 1
    
    for i in range(n):
        while j >= 0 and mat[i][j] == 1:
            j -= 1
        if m - j - 1 > max_ones:
            max_ones = m - j - 1
            max_index = i
    
    return max_index

def main():
    # Get the matrix dimensions from the user
    n = int(input("Enter the number of rows: "))
    m = int(input("Enter the number of columns: "))
    
    # Get the matrix elements from the user
    mat = []
    print("Enter elements of the matrix (in sorted order, 0s and 1s only):")
    for i in range(n):
        row = []
        for j in range(m):
            num = int(input(f"Enter element ({i + 1}, {j + 1}): "))
            if num not in [0, 1]:
                print("Error: Only 0s and 1s are allowed.")
                return
            row.append(num)
        mat.append(row)
    
    # Find the index of the row with maximum 1s
    result = find_max_ones_row(mat)
    
    # Print the result
    if result == -1:
        print("No row contains 1s.")
    else:
        print(f"The index of the first row with maximum 1s is: {result}")

if __name__ == "__main__":
    main()
```
