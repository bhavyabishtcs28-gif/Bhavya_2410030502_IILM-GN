```python
''' Performing operations on an array '''

import heapq

def min_operations(arr):
    """
    Calculate the minimum number of operations required to make the sum of array elements
    less than or equal to half of the original sum.
    
    :param arr: List of numbers
    :return: Minimum number of operations
    """
    total_sum = sum(arr)
    target_sum = total_sum / 2
    
    # Create a max heap to store elements
    max_heap = [-x for x in arr]
    heapq.heapify(max_heap)
    
    operations = 0
    while total_sum > target_sum:
        # Extract the largest element from the heap
        largest = -heapq.heappop(max_heap)
        
        # Reduce the largest element by half
        reduced = largest / 2
        total_sum -= largest - reduced
        
        # Push the reduced element back into the heap
        heapq.heappush(max_heap, -reduced)
        
        operations += 1
    
    return operations

def main():
    # Get user input
    n = int(input("Enter the number of elements: "))
    arr = []
    for i in range(n):
        arr.append(float(input(f"Enter element {i+1}: ")))
    
    # Calculate and display the result
    min_ops = min_operations(arr)
    print(f"Minimum operations required: {min_ops}")

if __name__ == "__main__":
    main()
```

```python
''' Working of a person (in array) '''

def min_people(arr):
    """
    Calculate the minimum number of people required to cover the entire working day.
    
    :param arr: List of ranges of working hours
    :return: Minimum number of people required
    """
    n = len(arr)
    intervals = []
    
    # Create intervals for each person
    for i in range(n):
        if arr[i] != -1:
            start = max(0, i - arr[i])
            end = min(n - 1, i + arr[i])
            intervals.append((start, end))
    
    # Sort intervals by start time
    intervals.sort(key=lambda x: x[0])
    
    # Initialize variables
    people = 0
    covered = 0
    i = 0
    
    while covered < n:
        # Find the interval that covers the most uncovered area
        next_covered = covered
        while i < len(intervals) and intervals[i][0] <= covered:
            next_covered = max(next_covered, intervals[i][1] + 1)
            i += 1
        
        # If no progress is made, it's not possible to cover the day
        if next_covered == covered:
            return -1
        
        covered = next_covered
        people += 1
        
        # If the entire day is covered, return the number of people
        if covered >= n:
            return people
    
    return people

def main():
    # Get user input
    n = int(input("Enter the number of people: "))
    arr = []
    for i in range(n):
        arr.append(int(input(f"Enter range for person {i+1}: ")))
    
    # Calculate and display the result
    min_people_required = min_people(arr)
    if min_people_required == -1:
        print("It's not possible to fully cover the day.")
    else:
        print(f"Minimum people required: {min_people_required}")

if __name__ == "__main__":
    main()
```

```python
''' PSE (Previous Smaller Element) '''

def previous_smaller_element(arr):
    """
    Calculate the Previous Smaller Element (PSE) for each element in the array.
    
    :param arr: List of integers
    :return: List of PSE for each element
    """
    n = len(arr)
    pse = [-1] * n
    stack = []
    
    for i in range(n):
        # Pop elements from the stack that are greater than or equal to arr[i]
        while stack and stack[-1] >= arr[i]:
            stack.pop()
        
        # If the stack is not empty, the top element is the PSE
        if stack:
            pse[i] = stack[-1]
        
        # Push the current element onto the stack
        stack.append(arr[i])
    
    return pse

def main():
    # Get user input
    n = int(input("Enter the number of elements: "))
    arr = []
    for i in range(n):
        arr.append(int(input(f"Enter element {i+1}: ")))
    
    # Calculate and display the result
    pse = previous_smaller_element(arr)
    print("Previous Smaller Element (PSE) for each element:")
    for i in range(n):
        print(f"{arr[i]} -> {pse[i]}")

if __name__ == "__main__":
    main()
```

```python
''' PSE (Previous Greater Element) '''

def previous_greater_element(arr):
    """
    Calculate the Previous Greater Element (PGE) for each element in the array.
    
    :param arr: List of integers
    :return: List of PGE for each element
    """
    n = len(arr)
    pge = [-1] * n
    stack = []
    
    for i in range(n):
        # Pop elements from the stack that are smaller than or equal to arr[i]
        while stack and stack[-1] <= arr[i]:
            stack.pop()
        
        # If the stack is not empty, the top element is the PGE
        if stack:
            pge[i] = stack[-1]
        
        # Push the current element onto the stack
        stack.append(arr[i])
    
    return pge

def main():
    # Get user input
    n = int(input("Enter the number of elements: "))
    arr = []
    for i in range(n):
        arr.append(int(input(f"Enter element {i+1}: ")))
    
    # Calculate and display the result
    pge = previous_greater_element(arr)
    print("Previous Greater Element (PGE) for each element:")
    for i in range(n):
        print(f"{arr[i]} -> {pge[i]}")

if __name__ == "__main__":
    main()

```

```python
''' Count of Subarrays (First element should be minimum element) '''

def count_subarrays(arr):
    """
    Count the number of subarrays where the first element is the minimum element.
    
    :param arr: List of integers
    :return: Number of subarrays
    """
    n = len(arr)
    count = 0
    
    for i in range(n):
        # Initialize the minimum element as the current element
        min_val = arr[i]
        
        # Iterate through the subarray starting from the current element
        for j in range(i, n):
            # Update the minimum element if a smaller element is found
            min_val = min(min_val, arr[j])
            
            # If the minimum element is the first element, increment the count
            if min_val == arr[i]:
                count += 1
    
    return count

def main():
    # Get user input
    n = int(input("Enter the number of elements: "))
    arr = []
    for i in range(n):
        arr.append(int(input(f"Enter element {i+1}: ")))
    
    # Calculate and display the result
    count = count_subarrays(arr)
    print(f"Number of subarrays: {count}")

if __name__ == "__main__":
    main()
```

```python
''' TO determine whether the array contains a 132 '''

def find_132_pattern(arr):
    """
    Determine whether the array contains a 132 pattern.
    
    :param arr: List of integers
    :return: True if a 132 pattern exists, False otherwise
    """
    n = len(arr)
    third = float('-inf')
    stack = []
    
    for i in range(n - 1, -1, -1):
        # If the current element is less than the third element, return True
        if arr[i] < third:
            return True
        
        # Pop elements from the stack that are smaller than the current element
        while stack and stack[-1] < arr[i]:
            third = stack.pop()
        
        # Push the current element onto the stack
        stack.append(arr[i])
    
    return False

def main():
    # Get user input
    n = int(input("Enter the number of elements: "))
    arr = []
    for i in range(n):
        arr.append(int(input(f"Enter element {i+1}: ")))
    
    # Calculate and display the result
    result = find_132_pattern(arr)
    print(f"132 pattern exists: {result}")

if __name__ == "__main__":
    main()
```

```python
''' Visibility of persons according to their height '''

def max_people_visible(arr):
    """
    Find the maximum number of people that any person can see.
    
    :param arr: List of heights of people
    :return: Maximum number of people visible
    """
    n = len(arr)
    max_visible = 0
    
    for i in range(n):
        # Initialize variables to store the number of people visible in both directions
        left_visible = 0
        right_visible = 0
        
        # Count people visible to the left
        for j in range(i - 1, -1, -1):
            if arr[j] < arr[i]:
                left_visible += 1
            elif arr[j] >= arr[i]:
                break
        
        # Count people visible to the right
        for j in range(i + 1, n):
            if arr[j] < arr[i]:
                right_visible += 1
            elif arr[j] >= arr[i]:
                break
        
        # Update the maximum visible count
        max_visible = max(max_visible, left_visible + right_visible + 1)
    
    return max_visible

def main():
    # Get user input
    n = int(input("Enter the number of people: "))
    arr = []
    for i in range(n):
        arr.append(int(input(f"Enter height of person {i+1}: ")))
    
    # Calculate and display the result
    max_visible = max_people_visible(arr)
    print(f"Maximum number of people visible: {max_visible}")

if __name__ == "__main__":
    main()

```

```python
''' Target sum elements '''

def combination_sum(n, k):
    """
    Find all valid combinations of k numbers that add up to n.
    
    :param n: Target sum
    :param k: Number of elements in each combination
    :return: List of combinations
    """
    def backtrack(start, path, target):
        if len(path) == k:
            if target == 0:
                result.append(path)
            return
        
        for i in range(start, 10):
            if i > target:
                break
            backtrack(i + 1, path + [i], target - i)
    
    result = []
    backtrack(1, [], n)
    return result

def main():
    # Get user input
    n = int(input("Enter the target sum: "))
    k = int(input("Enter the number of elements: "))
    
    # Calculate and display the result
    combinations = combination_sum(n, k)
    print("Combinations:")
    for combination in combinations:
        print(combination)

if __name__ == "__main__":
    main()
```

```python
''' Maximum and Minimum absolute difference '''

def max_min_diff(arr, k):
    """
    Select k elements from the array with maximum minimum absolute difference.
    
    :param arr: List of integers
    :param k: Number of elements to select
    :return: Maximum possible minimum difference
    """
    n = len(arr)
    arr.sort()
    
    # Define the search range for the minimum difference
    low = 0
    high = arr[-1] - arr[0]
    
    # Perform binary search to find the maximum minimum difference
    while low < high:
        mid = (low + high + 1) // 2
        count = 1
        prev = arr[0]
        
        # Check if it's possible to select k elements with minimum difference mid
        for i in range(1, n):
            if arr[i] - prev >= mid:
                count += 1
                prev = arr[i]
        
        # Update the search range
        if count >= k:
            low = mid
        else:
            high = mid - 1
    
    return low

def main():
    # Get user input
    n = int(input("Enter the number of elements: "))
    arr = []
    for i in range(n):
        arr.append(int(input(f"Enter element {i+1}: ")))
    k = int(input("Enter the number of elements to select: "))
    
    # Calculate and display the result
    max_min_diff_val = max_min_diff(arr, k)
    print("Maximum possible minimum difference:", max_min_diff_val)

if __name__ == "__main__":
    main()
```

```python
''' Count of elements in an array that are less than or equal to the value of x '''

def count_less_than_or_equal(arr, x):
    """
    Count the number of elements in the array that are less than or equal to x.
    
    :param arr: List of integers
    :param x: Target value
    :return: Number of elements less than or equal to x
    """
    n = len(arr)
    
    # Find the pivot index
    left, right = 0, n - 1
    while left < right:
        mid = (left + right) // 2
        if arr[mid] > arr[right]:
            left = mid + 1
        else:
            right = mid
    
    pivot = left
    
    # Perform binary search in the appropriate half
    if x >= arr[pivot] and x <= arr[n - 1]:
        left, right = pivot, n - 1
    else:
        left, right = 0, pivot - 1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] <= x:
            left = mid + 1
        else:
            right = mid - 1
    
    return left

def main():
    # Get user input
    n = int(input("Enter the number of elements: "))
    arr = []
    for i in range(n):
        arr.append(int(input(f"Enter element {i+1}: ")))
    x = int(input("Enter the target value: "))
    
    # Calculate and display the result
    count = count_less_than_or_equal(arr, x)
    print("Number of elements less than or equal to", x, ":", count)

if __name__ == "__main__":
    main()

```
